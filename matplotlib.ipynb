```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
fig, (ax1,ax2)=plt.subplots(1,2, figsize=[12,5],facecolor="brown",edgecolor='b')
fig.suptitle("Example Subplots")
ax1.plot([1,2,3,4,5,6,7],[32,26,43,41,16,37,29],color='r')
ax1.set_title("class 1")
ax1.set_xlabel("roll_no")
ax1.set_ylabel("marks out of 50")
ax2.plot([1,2,3,4,5,6,7],[48,32,40,44,36,27,21],color='g')
ax2.set_title("class 2")
ax2.set_xlabel("roll_no")
ax2.set_ylabel("marks out of 50")
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
fig, (ax1,ax2,ax3)=plt.subplots(1,3, figsize=[12,5],facecolor="brown",edgecolor='b')
fig.suptitle("Example Subplots")
ax1.plot([1,2,3,4,5,6,7],[32,26,43,41,16,37,29],color='r')
ax1.set_title("class 1")
ax1.set_xlabel("roll_no")
ax1.set_ylabel("marks out of 50")
ax2.plot([1,2,3,4,5,6,7],[48,32,40,44,36,27,21],color='g')
ax2.set_title("class 2")
ax2.set_xlabel("roll_no")
ax2.set_ylabel("marks out of 50")
ax3.plot([1,2,3,4,5,6,7],[48,30,41,48,31,45,21],color='b')
ax3.set_title("class 3")
ax3.set_xlabel("roll_no")
ax3.set_ylabel("marks out of 50")
```

A Line plot  visualizes the information as a series  of data points called markers connected by straight line  segments.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
date={
    "days":['Monday','tuesday','wednesday','Thursday','friday','saturday','sunday'],'Week1':[27,28,32,33,33,31,29],'Week2':[29,30,31,32,33,34,31],'Week3':[29,28,25,27,31,30,20]
}
df=pd.DataFrame(date)
plt.plot(df["days"],df["Week1"])
plt.plot(df["days"],df["Week2"])
plt.plot(df["days"],df["Week3"])
plt.title("Line plot of week 1 Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend(["week1","week2","week3"])
plt.show()
```
